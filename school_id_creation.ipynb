```python
# import dependencies
import pandas as pd
```

```python
# read in players data
players_df = pd.read_csv('resources/players_school_and_size.csv', encoding='windows-1252')
players_df
```

```python
players_df['college'] = players_df['college'].fillna('International/High School Prospect')
players_df['college'] = players_df['college'].astype('string')
players_df
```

```python
players_df.info()
```

```python
# Find Unique Schools
schools = pd.DataFrame(players_df['college'].unique(), columns=['college']).sort_values('college')
schools = schools.reset_index(drop=True) 
schools
```

```python
# change college to string data
schools['college'] = schools['college'].astype('string')
schools
```

```python
# Assign IDs to Schools
schools['School_ID'] = ''

id_num = 0

for row in schools['School_ID']:
    if id_num <= 9:
        schools['School_ID'][id_num] = ('000' + str(id_num))
    elif id_num <= 99:
        schools['School_ID'][id_num] = ('00' + str(id_num))
    elif id_num <= 999:
        schools['School_ID'][id_num] = ('0' + str(id_num))
    id_num += 1    
schools
```

```python
schools.info()
```

```python
# Merge the School IDs with Player Data
players_df = players_df.merge(schools, on=['college'], how='left')
players_df
```

```python
# Write the dataframe to a csv
players_df.to_csv('transition_files/schoolID_player.csv', index = False, encoding = 'UTF-8')
```
