```python
# import dependencies
import pandas as pd
```

```python
# Read in Team Table
team_table = pd.read_csv('database_files/team_table.csv', encoding='windows-1252')
team_table
```

```python
# read in NBA awards transition file
nba_results = pd.read_csv('transition_files/nba_awards.csv', encoding='UTF-8')
nba_results
```

```python
# Setup ID Table
team_ids = pd.DataFrame(team_table['Franchise'])
team_ids['TeamID'] = team_table['Abbreviation']
team_ids = team_ids.rename(columns={"Franchise": "Champion"})
team_ids
```

```python
# Fill NA
nba_results['Champion'] = nba_results['Champion'].fillna('Playoffs Unfinished')

# Set Champion to Upper
for i in range(71):
    nba_results['Champion'][i] = nba_results['Champion'][i].upper()
nba_results
```

```python
# Merge Team IDs into NBA Results
nba_results = nba_results.merge(team_ids, on='Champion', how='left')
nba_results
```

```python
# Add Awards Not Issued
nba_results = nba_results.fillna('AWARD NOT ISSUED')
nba_results
```

```python
# adjust Season to year ending
for i in range(71):
    front = nba_results['Season'][i][0:2]
    back = nba_results['Season'][i][5:7]
    year = int(str(front)+str(back))
    nba_results['Season'][i] = year
```

```python
# Final Table View
nba_results
```

```python
# Team ID Check, Award not issued should only be 1
nba_results['TeamID'].value_counts()
```

```python
# Remove recent years to match players table
nba_results=nba_results[nba_results['Season'] != 2020]
nba_results=nba_results[nba_results['Season'] != 2019]
nba_results
```

```python
# Remove Years Where Awards are not Issued
nba_results = nba_results.iloc[0:63]
nba_results
```

```python
# Write the dataframe to a csv
nba_results.to_csv('database_files/nba_awards.csv', index = False, encoding = 'UTF-8')
```
