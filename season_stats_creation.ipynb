```python
# import dependencies
import pandas as pd
```

```python
# read in season stats data
season_stats_df = pd.read_csv('resources/seasons_stats.csv', encoding='windows-1252')
season_stats_df
```

```python
# analyze data types for counting stats
season_stats_df.info()
```

```python
# Check for empty rows
season_stats_df['Year'].isna().sum()
```

```python
# Drop Empty Rows
season_stats_df.dropna(subset=['Year'], inplace=True)
season_stats_df = season_stats_df.reset_index(drop=True)
season_stats_df
```

```python
# Drop Null Columns
season_stats_df = season_stats_df.drop(['blanl', 'blank2'], axis=1)
season_stats_df
```

```python
# Remove TOT teams, as we would like to calculate totals on our own
season_stats_df = season_stats_df[season_stats_df['Tm'] != 'TOT']
season_stats_df
```

```python
season_stats_df.info()
```

```python
# read in player dataset
players_df = pd.read_csv('database_files/player_dataset.csv', encoding='UTF-8')
players_df
```

```python
# create player id table
player_id_table = pd.DataFrame(players_df['name'])
player_id_table['player_id'] = players_df['player_id']
player_id_table = player_id_table.rename(columns={"name": "Player"})
player_id_table
```

```python
# Merge Player ID Table with Season Stats
nba_stats_df = season_stats_df.merge(player_id_table, on=['Player'], how='inner').sort_values(by='Year', axis=0) \
                .reset_index(drop=True)
nba_stats_df
```

```python
# Write Nba Season Stats to CSV
nba_stats_df.to_csv('database_files/nba_season_stats.csv', index = False, encoding = 'UTF-8')
```
