```python
# import dependencies
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
import openpyxl
```

```python
# read in nba awards data
nba_awards_df = pd.read_csv('resources/nba_results.csv', encoding='windows-1252')
nba_awards_df
```

```python
# chromedriver setup
executable_path = {'executable_path': 'chromedriver/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# html page setup
url = 'https://www.basketball-reference.com/leagues/'
browser.visit(url)
```

```python
# use beautiful soup to find all the table data
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

datagrid = soup.find('table')

mvp_player_page = ['/players/a/antetgi01.html']
roy_player_page = []
pts_player_page = []
trb_player_page = []
ast_player_page = []
ws_player_page = []


for tr in datagrid.find_all('tr')[2:]:
    try:
        for td in tr.find_all('td')[0]: # League Data
            print(td.text)
    except:
        print('No data')
    try:
        for td in tr.find_all('td')[1]: # Champion Data
            print(td.text)
    except:
        print('No data')    
    try:
        for td in tr.find_all('td')[2]: # MVP Data
            print(td['href'])
            mvp_player_page.append(td['href'])
    except:
        print('No data')
    try:
        for td in tr.find_all('td')[3]: # ROY Data
            print(td['href'])
            roy_player_page.append(td['href'])
    except:
        print('No data')
    try:
        for td in tr.find_all('td')[4]: # PTS Data
            print(td['href'])
            pts_player_page.append(td['href'])
    except:
        print('No data')
    try:
        for td in tr.find_all('td')[5]: # TRB Data
            print(td['href'])
            trb_player_page.append(td['href'])
    except:
        print('No data')
    try:
        for td in tr.find_all('td')[6]: # AST Data
            print(td['href'])
            ast_player_page.append(td['href'])
    except:
        print('No data')
    try:
        for td in tr.find_all('td')[7]: # WS Data
            print(td['href'])
            ws_player_page.append(td['href'])
    except:
        print('No data')
```

```python
mvp_player_page
```

```python
# Transform Player Pages into Player Ids
back = '.html'

for i in range(83):
    try:
        # MVP Page
        mvp_player_page[i] = mvp_player_page[i][11:]
        mvp_player_page[i] = mvp_player_page[i].replace(back, '')
    except (IndexError):
        pass
    try:
        # ROY Page
        roy_player_page[i] = roy_player_page[i][11:]
        roy_player_page[i] = roy_player_page[i].replace(back, '')
    except (IndexError):
        pass
    try:
        # PTS Page
        pts_player_page[i] = pts_player_page[i][11:]
        pts_player_page[i] = pts_player_page[i].replace(back, '')
    except (IndexError):
        pass
    try:
        # TRB Page
        trb_player_page[i] = trb_player_page[i][11:]
        trb_player_page[i] = trb_player_page[i].replace(back, '')
    except (IndexError):
        pass
    try:
        # AST Page
        ast_player_page[i] = ast_player_page[i][11:]
        ast_player_page[i] = ast_player_page[i].replace(back, '')
    except (IndexError):
        pass
    try:
        # WS Page
        ws_player_page[i] = ws_player_page[i][11:]
        ws_player_page[i] = ws_player_page[i].replace(back, '')
    except (IndexError):
        pass
```

```python
# Creat dataframe with awards using player pages as id's 
player_pages = pd.DataFrame.from_dict({
    'mvp_player_id': pd.Series(mvp_player_page),
    'roy_player_id': pd.Series(roy_player_page),
    'pts_player_id': pd.Series(pts_player_page),
    'trb_player_id': pd.Series(trb_player_page),
    'ast_player_id': pd.Series(ast_player_page),
    'ws_player_id': pd.Series(ws_player_page),
}, orient='index').T 
player_pages
```

```python
# Merge the two dataframes, with a left join using index as the default 'on'
nba_results = nba_awards_df.join(player_pages, how='left')
nba_results
```

```python
# Remove ABA and BAA Titles
nba_results = nba_results[nba_results['Lg'] != 'ABA']
nba_results = nba_results[nba_results['Lg'] != 'BAA']
nba_results
```

```python
# Write the dataframe to a csv
nba_results.to_csv('transition_files/nba_awards.csv', index = False, encoding = 'UTF-8')
```

```python
# Read in Team Table
team_table = pd.read_csv('database_files/team_table.csv', encoding='windows-1252')
team_table
```

```python
# Setup ID Table
team_ids = pd.DataFrame(team_table['Franchise'])
team_ids['ID'] = team_table['Abbreviation']
team_ids = team_ids.rename(columns={"Franchise": "Champion"})
team_ids
```

```python
# Fill NA
nba_results['Champion'] = nba_results['Champion'].fillna('Playoffs Unfinished')

# Set Champion to Upper
for i in range(45):
    nba_results['Champion'][i] = nba_results['Champion'][i].upper()
nba_results
```

```python
nba_results['Champion'][81] = 'BALTIMORE BULLETS (1948-1955)'
```

```python
# Merge Team IDs into NBA Results
nba_results = nba_results.merge(team_ids, on='Champion', how='left')
nba_results = nba_results.fillna('AWARD NOT ISSUED')
nba_results
```

```python
# Write the dataframe to a csv
nba_results.to_csv('database_files/nba_awards.csv', index = False, encoding = 'UTF-8')
```
