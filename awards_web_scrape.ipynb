```python
# import dependencies
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
# read in nba awards data
nba_awards_df = pd.read_excel('resources/nba_results.xlsx')
nba_awards_df
```

```python
# chromedriver setup
executable_path = {'executable_path': 'chromedriver/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# html page setup
url = 'https://www.basketball-reference.com/leagues/'
browser.visit(url)
```

```python
# use beautiful soup to find all the table data
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

table_body = soup.find('tbody')
table_data = table_body.find_all('tr')
table_data.pretty()

```

```python
# Demo Scraping to find url 
player_links = []
# Use Beautiful Soup's find() method to navigate and retrieve attributes
for data in table_data:
    try:
        if data.text == 'ABA':
            continue
        link = data.find('a')
        href = link['href']
        if href[1:8] == 'players':
            player_links.append(href)
        print('-----------')
        if href[1:8] == 'leagues':
            print(href[13:17] + ' Season')
        print(data.text)
        print('https://www.basketball-reference.com' + href)
        if int(href[13:17]) <= 1970:
            break
    except:
        print('Scraping Attempt Failed.')
        print('Processing...')
```
