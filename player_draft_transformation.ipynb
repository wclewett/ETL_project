```python
# import dependencies
import pandas as pd
```

```python
# read in players data
draft_df = pd.read_csv('resources/players_draft.csv', encoding='UTF-8')
draft_df
```

```python
draft_df = draft_df.rename(columns={"Player": "name"})
draft_df 
```

```python
#Read a player dataset
player= pd.read_csv('database_files/player_dataset.csv', encoding='UTF-8')
player
```

```python
#Getting name, school_id and player_id as a primary key
name_df = pd.DataFrame(player['name'])
name_df['College'] = player['School']
name_df['School_ID'] = player['School_ID']
name_df['Player_ID'] = player['player_id']
name_df
```

```python
#Merge player on draft_df by using left join
draft_df=draft_df.merge(name_df, on='name', how='left')
draft_df
```
