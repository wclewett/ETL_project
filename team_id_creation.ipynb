```python
# import dependencies
import pandas as pd
```

```python
# read in season stats
season_stats = pd.read_csv('resources/seasons_stats.csv', encoding='windows-1252')
season_stats
```

```python
# There are entirely blank rows within the dataset, we will drop these before proceeding
season_stats = season_stats[season_stats['Year'].notna()]
season_stats['Year'] = season_stats['Year'].astype(int)
season_stats
```

```python
# We will also drop TOT rows, these appear when a player has been traded during the year and consequently plays for 
# more than one team in a season
season_stats = season_stats[season_stats['Tm'] != 'TOT']
season_stats
```

```python
# read in team abbreviations
team_abbreviations = pd.read_csv('resources/team_abbreviations.csv', encoding='windows-1252')
team_abbreviations['Franchise'] = team_abbreviations['Franchise'].astype('string')
team_abbreviations['Abbreviation'] = team_abbreviations['Abbreviation'].astype('string')
team_abbreviations
```

```python
# read in franchise data
franchise_data = pd.read_csv('resources/franchise_data.csv', encoding='windows-1252')
franchise_data['Franchise'] = franchise_data['Franchise'].astype('string')
franchise_data
```

```python
# adjust years to match year ending
for i in range(111):
    front = franchise_data['From'][i][0:2]
    back = franchise_data['From'][i][5:7]
    year = int(str(front)+str(back))
    franchise_data['From'][i] = year
    
for i in range(111):
    front = franchise_data['To'][i][0:2]
    back = franchise_data['To'][i][5:7]
    year = int(str(front)+str(back))
    franchise_data['To'][i] = year

franchise_data
```

```python
# Remove non-NBA teams
franchise_data = franchise_data[franchise_data['Lg'] != 'BAA']
franchise_data = franchise_data[franchise_data['Lg'] != 'ABA']
franchise_data = franchise_data[franchise_data['Lg'] != 'NBA/ABA']
franchise_data = franchise_data.reset_index(drop=True)
franchise_data
```

```python
# Set the Franchise name to match the syntax in the Team Abbreviations Table
for i in range(65):
    franchise_data['Franchise'][i] = franchise_data['Franchise'][i].upper()
franchise_data
```

```python
# First Merge the Team Abbrevations with the Franchise Data
team_table = franchise_data.merge(team_abbreviations, on=['Franchise'], how='left')
team_table
```

```python
# Check for Duplicates
team_table['Abbreviation'].value_counts().max()
```

```python
team_table = team_table.rename(columns={"From": "Team_Start", "To": "Team_End", "W/L%": "W_L_PCT"})
team_table['Div'] = team_table['Div'].astype(int) 
team_table
```

```python
# Write the dataframe to a csv
team_table.to_csv('database_files/team_table.csv', index = False, encoding = 'UTF-8')
```
