```python
# import dependencies
import pandas as pd
```

```python
# read in players data
players_df = pd.read_csv('resources/players_data.csv', encoding='windows-1252')
players_df
```

```python
# Transform Player Pages into Player Ids
back = '.html'


for i in range(4576):
    try:
        # URL Column
        players_df['url'][i] = players_df['url'].copy()[i][11:]
        players_df['url'][i] = players_df['url'].copy()[i].replace(back, '')
    except (IndexError):
        pass

players_df = players_df.rename(columns={"url": "player_id"})
players_df
```

```python
# Years Active
players_df['NBA Years Active'] = players_df['active_to'] - players_df['active_from'] + 1
players_df
```
