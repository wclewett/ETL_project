```python
# import dependencies
import pandas as pd
```

```python
# read in players data
players_df = pd.read_csv('resources/players_data.csv', encoding='windows-1252')
players_df
```

```python
# Transform Player Pages into Player Ids
back = '.html'


for i in range(4576):
    try:
        # URL Column
        players_df['url'][i] = players_df['url'].copy()[i][11:]
        players_df['url'][i] = players_df['url'].copy()[i].replace(back, '')
    except (IndexError):
        pass

players_df = players_df.rename(columns={"url": "player_id"})
players_df
```

```python
# Years Active
players_df['NBA Years Active'] = players_df['active_to'] - players_df['active_from'] + 1
players_df
```

```python
# Change Missing College NaN's to International/High School Prospect
players_df['college'] = players_df['college'].fillna('International/High School Prospect')
players_df
```

```python
# Read in SchoolID_Player Dataset
school_id = pd.read_csv('transition_files/schoolID_player.csv', encoding='windows-1252')
school_id
```

```python
school_df = pd.DataFrame(school_id['name'])
school_df['college'] = school_id['college']
school_df['School_ID'] = school_id['School_ID']
school_df
```

```python
# Merge the Dataset with the Player/School ID Dataset
players_df = players_df.merge(school_df, on='name', how='left')
players_df
```

```python
# test for duplicates
duplicate_test_bool = players_df['player_id'].value_counts() > 1
duplicate_test_bool
```

```python
# Survey Duplicates 
players_df[players_df['player_id'] == 'smithch01']
```

```python
# Drop Duplicates
players_df = players_df.drop_duplicates(subset=['player_id'])
players_df
```

```python
# Drop Duplicated Columns
players_df = players_df.drop(['college_x'], axis=1)
players_df
```

```python
players_df = players_df.reset_index(drop=True)
players_df
```

```python
# Convert height to inches

for i in range(4576):
    feet = int(players_df['height'][i][1])
    inches = int(players_df['height'][i][3:5])
    players_df['height'][i] = (feet * 12) + inches

players_df
```
